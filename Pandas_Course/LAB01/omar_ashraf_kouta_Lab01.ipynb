## Question 3

```python
df = pd.read_csv("needed_files/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df = df.set_index("customerID")
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"].isnull().sum() / len(df)
```

* No missings in Total Charges

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df["TotalCharges"].isnull().sum() / len(df)
```

* Now we have some missing values

* I don't know why ? I saw unique values before conversion and there is nothing unusual.

```python
df["TotalCharges"].fillna(df["TotalCharges"].mean(), inplace=True)
print(df["TotalCharges"].isnull().sum() / len(df))
```

```python
cats = df.select_dtypes(include=['object']).columns.tolist()

cats = [col for col in cats if col not in ['Contract', 'Churn']]

cats
```

```python
df_ohe = pd.get_dummies(df, columns=cats)
df_ohe.head()
```

```python
df_ohe["Contract"].value_counts()
```

```python
df_ohe.replace({"Contract": {"Month-to-month": 1, "One year": 2, "Two year": 3}}, inplace=True)
df_ohe["Contract"].value_counts()
```

```python
df_ohe["Contract"].isnull().sum()
```

```python
df_ohe.duplicated().sum()
```

```python
df_ohe = df_ohe[~df_ohe.duplicated()]
```

```python
df_ohe.shape
```

```python
df_ohe["Churn"].replace({"Yes": 1, "No": 0}, inplace=True)
df_ohe["Churn"].value_counts()
```

```python
df_ohe["Churn"] = pd.to_numeric(df_ohe["Churn"], errors='coerce')
```

```python
df_ohe["Churn"].isnull().sum()
```

```python
target = df_ohe["Churn"]
features = df_ohe.drop("Churn", axis=1)
```

```python
numerical_cols = features.select_dtypes(include=['int64', 'float64']).columns
features[numerical_cols] = (features[numerical_cols] - features[numerical_cols].mean()) / features[numerical_cols].std()
features.head()
```

```python
features.to_csv("features.csv")
target.to_csv("target.csv")
```
