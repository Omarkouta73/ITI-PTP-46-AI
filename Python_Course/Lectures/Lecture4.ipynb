### Count the number of objects created

```python
# method 1

n_objects = 0
class Student:
    __slots__ = ("name", "email", "salary")
    def __init__(self, name, email, salary):
        global n_objects
        n_objects+=1
        self.name = name
        self.email = email
        self.salary = salary
        print(self) # self --> current obj == this
    
    # instance method
    def displayStudent(self):
        pass

s = Student("Omar", "omar@gmail.com", 10000)
s2 = Student("mohamed", "mo@gmail.com", 12000)
print(n_objects)
print(sys.getsizeof(s))
```

```python
# method 2

class Student:
    __slots__ = ("name", "email", "salary")
    n_objects = 0 # correct
    def __init__(self, name, email, salary):
        self.name = name
        self.email = email
        self.salary = salary
        Student.n_objects += 1 # like static variable
        print(self) # self --> current obj == this
    
    # instance method
    def displayStudent(self):
        print("inner fun: ", self.n_objects)


# Student.n_objects = 0 # correct
## it is smth like Employee.work_days = 5 this way i won't set work_days for each instance
s = Student("Omar", "omar@gmail.com", 10000)
s2 = Student("mohamed", "mo@gmail.com", 12000)
print("out: ", s2.n_objects)
print("fun: ", s2.displayStudent())
print(sys.getsizeof(s))
```

```python
class Student:
    # __slots__ = ("name", "email", "salary")
    n_objects = 0 # correct
    def __init__(self, name, email, salary):
        self.name = name
        self.email = email
        self.salary = salary
        Student.n_objects += 1 # like static variable
        # self.__class__.n_objects +=1
        print(self) # self --> current obj == this
    
    # instance method
    def displayStudent(self):
        print("inner fun: ", self.n_objects)
    
    @classmethod # https://refactoring.guru/design-patterns/decorator
    def create_default_object(cls):
        return cls("omar", "omar@gmail.com", 1000)
    
    @staticmethod
    def calc_tax(salary): # this helper function doesn't depend on instance or the class
        return salary * 0.8

s1 = Student.create_default_object()
print(Student.calc_tax(1000))
print(s1.calc_tax(1000))
print(s1.calc_tax(s1.salary))
print(s1.__dict__)
```
