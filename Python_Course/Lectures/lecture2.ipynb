```python
import numpy as np

a = np.array([1])

print(a.nbytes)  # 12 bytes (3 elements × 4 bytes each)
```

```python
import numpy as np

a = np.array([1], dtype=np.int8)

print(a.nbytes)  # 12 bytes (3 elements × 4 bytes each)
```

```python
import numpy as np

a = np.array([1], dtype=np.float32)
print(a.nbytes)
```

```python
a = np.array([1])
b = a
b[0] = 2

print(hex(a.ctypes.data))
print(hex(b.ctypes.data))
print("-"*20)
# print(a[0], "at", hex(id(a[0])))
print(a, "at", hex(id(a)))
print(b, "at", hex(id(b)))
```

## Functions without unknown number of parameters

```python
# function accept zero or more => use regex

def askforstudents(*students): # * -> zero or more
    print(students)

askforstudents("omar", "ahmed", "mohamed")
```

```python
def intromyself(info):
    print(info)

print(name="omar", age=22, gpa=3.2)
```

```python
def intromyself(**info):
    print(info)

intromyself(name="omar", age=22, gpa=3.2)
```
