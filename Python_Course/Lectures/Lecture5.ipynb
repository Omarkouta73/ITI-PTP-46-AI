```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, salary):
        self.salary = salary

emp = Employee()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, salary):
        self.salary = salary

emp = Employee(1000)
```

```python
# to call parent constructor --> call it explicitly

class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    # override constructor
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

emp = Employee("OMAR", 1000)
```
