Why Python int is ~28+ bytes

Python’s integer is not a primitive.
It’s a full object, with all the “Python object overhead” included.

Here’s what’s inside every Python int object on CPython:

1. Reference count (8 bytes)
```
x = 5
y = x
```
Now 5 has refcount = 2. Why? x points to the object 5 and y also points to the same object 5

2. Pointer to the int type object (8 bytes)
* Every object in Python stores a pointer to its class (type).
* Because Python is dynamically typed: 
    * it means that the type of a variable is determined at runtime, not in advance. In other words:
        * You don’t need to declare a type when creating a variable.
        * The same variable can hold values of different types at different times.

Type checking happens while the program runs, not when you write the code.

3. Various object flags / memory alignment (~8 bytes)
4. The integer value itself

Python uses arbitrary precision, meaning it can store infinitely big integers, not limited to 32 bits.

Storing something like 1000000000000000000000000000000 is legal.

So the storage for the value grows.
For small ints (like 0, 1, 2), it uses around 8 bytes

````python
x = 5
y = x

print(x, " at ", hex(id(x)))
print(y, " at ", hex(id(y)))
````

2. Small integers are cached:
* Integers from −5 to 256 are pre-allocated by Python, so their addresses don’t change across uses.

````python
x = 10
y = 10

print(x, " at ", hex(id(x)))
print(y, " at ", hex(id(y)))

print("="*20)

x = 11
y = 11

print(x, " at ", hex(id(x)))
print(y, " at ", hex(id(y)))

print("="*20)

x = 12
y = 12

print(x, " at ", hex(id(x)))
print(y, " at ", hex(id(y)))

print("="*20)

x = 300
y = 300

print(x, " at ", hex(id(x)))
print(y, " at ", hex(id(y)))
````

````python
import ctypes

x = 5

# get pointer to Python object x
addr = id(x)

# interpret that memory as a PyObject
refcount = ctypes.c_long.from_address(addr)
type_ptr = ctypes.c_void_p.from_address(addr + ctypes.sizeof(ctypes.c_long))

print("refcount:", refcount.value)
print("type address:", hex(type_ptr.value))
print("type object:", x.__class__)
````
