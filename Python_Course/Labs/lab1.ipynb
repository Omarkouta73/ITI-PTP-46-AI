## Count the vowels in a string

```python
vowels = "aeiouAEIOU"
text = "Hello World"
n = 0
for ch in text:
    if ch not in vowels:
        continue
    n+=1

print(n)
```

## Count 'iti' in a string

```python
text = "I love iti . iti is the best educational orginization"

print(text.count("iti"))
```

## Remove vowels from a text

```python
vowels = "aeiouAEIOU"
text = "My name is Omar Kouta"

for ch in text:
    if ch not in vowels:
        continue
    text = text.replace(ch, "")

print(text)
```

## Print the locations of i

```python
text = "i love iti"
n = ""
for i in range(len(text)):
    if text[i] == 'i':
        n = n + str(i) + ", "
    else:
        continue

print(n)
```

## Multiplications

```python
x = input("Enter integer number: ")

# 1 => 1
# 2 => (1x1 , 2x1, 2x2)
# 3 => (1x1, 2x1, 2x2, 3x1, 3x2, 3x3)
s = ""
if x.isdigit():
    x = int(x)
    for i in range(1, x+1):
        for j in range(1, i+1):
            s+=str(i*j)+", "
    print(s)

else:
    print("your input is not integer. Please re-enter the number...")
```

## Mario pyramid

```python
x = input("Enter integer number: ")

if x.isdigit():
    x = int(x)
    for i in range(1, x+1):
        print('*'*i)

else:
    print("your input is not integer. Please re-enter the number...")
```

```python
x = input("Enter integer number: ")

if x.isdigit():
    x = int(x)
    for i in range(1, x+1):
        print(" "*(x-i) + "*"*(i))

else:
    print("your input is not integer. Please re-enter the number...")
```

## Fill an array of 5 elements then sort it desc and asc then display the output

```python
x = []
for i in range(5):
    n = input(f"Enter element {i}: ")
    if n.isdigit():
        x.append(int(n))
    else:
        print(False)
        x.append(n)

x.sort(reverse=True)
print(f"Desc: {x}")
x.sort()
print(f"Asc: {x}")
```

## Multiplications with list

```python
x = input("Enter integer number: ")

# 1 => 1
# 2 => (1x1 , 2x1, 2x2)
# 3 => (1x1, 2x1, 2x2, 3x1, 3x2, 3x3)
s = []
if x.isdigit():
    x = int(x)
    for i in range(1, x+1):
        for j in range(1, i+1):
            s.append(i*j)
    print(s)

else:
    print("your input is not integer. Please re-enter the number...")
```
