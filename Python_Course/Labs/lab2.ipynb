## Write a function that accepts two arguments (length, start) to generate an array of a specific length filled with integer numbers increased by one from start.

```python
def generateArray(length, start):
    return list(range(start, length+1, 1))

print(generateArray(4, 1))
```

## Write a function that takes a number as an argument and if the number divisible by 3 return "Fizz" and if it is divisible by 5 return "buzz" and if is is divisible by both return "FizzBuzz"

```python
def fizzerbuzzer(n):
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return None

print (fizzerbuzzer(15))
```

## Write a function which has an input of a string from user then it will return the same string reversed

```python
def reverseString1(s):
    return ''.join(reversed(s))


def reverseString2(s):
    return s[::-1]


print(reverseString1("hello"))
print(reverseString2("hello"))
```

● Write a program which repeatedly reads numbers until the user
enters “done”.
○ Once “done” is entered, print out the total, count, and
average of the numbers.
○ If the user enters anything other than a number, detect their
mistake, print an error message and skip to the next number

```python
arr = []


while True:
    inp = input("Enter Number: ")
    if inp.isdigit():
        arr.append(int(inp))
    elif inp.isalpha() and inp == "done":
        break
    else:
        continue


print(f"""
    Total: {sum(arr)}
    Count: {len(arr)}
    AVG: {sum(arr)/len(arr) if len(arr) > 0 else 0}
""")
```

● Word guessing game (hangman)
○ A list of words will be hardcoded in your program, out of
which the interpreter will
○ choose 1 random word.
○ The user first must input their names
○ Ask the user to guess any alphabet. If the random word
contains that alphabet, it
○ will be shown as the output(with correct placement)
○ Else the program will ask you to guess another alphabet.
○ Give 7 turns maximum to guess the complete word.

```python
import random

def hangman():
    words = ["python", "mansoura", "programming", "iti", "linux", "disk"]

    word = random.choice(words)
    print(word)

    word_letters = set(word)

    name = input("Enter your name: ")
    print(f"Hello {name}")
    
    turns = 7
    
    display = ["_" for ch in word]
    
    while turns > 0:
        print("Word: ", " ".join(display))
        guess = input("Guess a letter: ").lower()
        
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter 1 alphabet")
            continue
        
        
        if guess in word_letters:
            for i, letter in enumerate(word):
                if letter == guess:
                    display[i] = guess
            print(f"Good work! Continue...")
        else:
            turns -= 1
            print(f"Try Again! '{guess}' is not in the word. Turns left: {turns}")
        
        if "_" not in display:
            print("Congratulations! The word is:", word)
            return
    
    print("Game Over! The word was:", word)

hangman()

# words = ["python", "Mansoura", "programming", "ITI", "Linux", "Disk"]
```
