# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:**
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
sample_sizes = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]

for n in sample_sizes:
    flips = np.random.randint(0, 2, size=n)    
    num_heads = np.sum(flips)
    prob = num_heads / n
    print(f"n = {n}, P(e) = {prob}")
    print("-" * 30)

n_max = 1000
flips = np.random.randint(0, 2, size=n_max)

cumulative_heads = np.cumsum(flips)
trials = np.arange(1, n_max + 1)
cumulative_prob = cumulative_heads / trials

plt.figure(figsize=(8, 5))
plt.plot(trials, cumulative_prob, 'b-', linewidth=1)
plt.axhline(y=0.5, color='r', linestyle='-', linewidth=2)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
for n in sample_sizes:
    rolls = np.random.randint(1, 7, size=n)    
    num_fives = np.sum(rolls == 5)    
    prob = num_fives / n
    
    print(f"n = {n}, P(e) = {prob}")
    print("-" * 30)

rolls = np.random.randint(1, 7, size=n_max)

cumulative_fives = np.cumsum(rolls == 5)
trials = np.arange(1, n_max + 1)
cumulative_prob = cumulative_fives / trials

plt.plot(trials, cumulative_prob, 'b-', linewidth=1)
plt.axhline(y=1/6, color='r', linestyle='-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.show()
```

3. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
#type your code here:
for n in sample_sizes:
    rolls = np.random.randint(1, 7, size=n)
    num_greater_than_3 = np.sum(rolls > 3)
    prob = num_greater_than_3 / n
    print(f"n = {n}, P(e) = {prob}")
    print("-" * 30)

rolls = np.random.randint(1, 7, size=n_max)

cumulative_gt3 = np.cumsum(rolls > 3)
trials = np.arange(1, n_max + 1)
cumulative_prob = cumulative_gt3 / trials

plt.plot(trials, cumulative_prob, 'b-', linewidth=1)
plt.axhline(y=0.5, color='r', linestyle='-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.show()
```

4. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
#type your code here:
for n in sample_sizes:
    die1 = np.random.randint(1, 7, size=n)
    die2 = np.random.randint(1, 7, size=n)
    sums = die1 + die2
    num_sum_five = np.sum(sums == 5)
    prob = num_sum_five / n
    print(f"n = {n}, P(e) = {prob}")
    print("-" * 30)

die1 = np.random.randint(1, 7, size=n_max)
die2 = np.random.randint(1, 7, size=n_max)
sums = die1 + die2

cumulative_sum5 = np.cumsum(sums == 5)
trials = np.arange(1, n_max + 1)
cumulative_prob = cumulative_sum5 / trials

plt.plot(trials, cumulative_prob, 'b-', linewidth=1)
plt.axhline(y=4/36, color='r', linestyle='-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.show()
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
#type your code here:

# simulate 1 million tests of two fair coin flips
flips = np.random.randint(0, 2, size=(1000000, 2))

# sums of all tests

sums = np.sum(flips, axis=1)

# proportion of tests that produced exactly two heads
prob = np.mean(sums == 2)

print(prob)
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

# simulate 1 million tests of three fair coin flips
flips = np.random.randint(0, 2, size=(1000000, 3))

# sums of all tests

sums = np.sum(flips, axis=1)

# proportion of tests that produced exactly one head
prob = np.mean(sums == 1)

print(prob)
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
#type your code here:


# simulate 1 million tests of three biased coin flips
# hint: use np.random.choice()
flips = np.random.choice([0, 1], size=(1000000, 3), p=[0.4, 0.6])

# sums of all tests
sums = np.sum(flips, axis=1)

# proportion of tests that produced exactly one head
prob = np.mean(sums == 1)

print(prob)
```

# Problem 4:Random Variables Statistical Tools
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
#type your code here:
rolls = np.random.randint(1, 7, size=1000)

sample_mean = np.mean(rolls)
sample_variance = np.var(rolls, ddof=1)
sample_std = np.std(rolls, ddof=1)

print(f"Sample mean: {sample_mean}")
print(f"Sample variance: {sample_variance}")
print(f"Sample standard deviation: {sample_std}")
```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparant_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
```

```python
plt.scatter(temperature, apparant_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

## Temperature and Apparent Temperature

```python
#type your code here:

temp_mean = np.mean(temperature)
apparent_mean = np.mean(apparant_temp)

covariance = np.mean((temperature - temp_mean) * (apparant_temp - apparent_mean))

print("Covariance: ", covariance)
print("-" * 40)

cov_matrix = np.cov(temperature.flatten(), apparant_temp.flatten())
print("np.cov(): \n", cov_matrix)
print("-" * 40)

temp_std = np.std(temperature)
apparent_std = np.std(apparant_temp)
correlation = covariance / (temp_std * apparent_std)

print("Correlation Cof: ", correlation)
print("-" * 40)

corr_matrix = np.corrcoef(temperature.flatten(), apparant_temp.flatten())
print("np.corrcoef(): \n", corr_matrix)
```

## Temperature and Humidity

```python
#type your code here:


humidity_mean = np.mean(humidity)

covariance = np.mean((temperature - temp_mean) * (humidity - humidity_mean))

print("Covariance: ", covariance)
print("-" * 40)

cov_matrix = np.cov(temperature.flatten(), humidity.flatten())
print("np.cov(): \n", cov_matrix)
print("-" * 40)

humidity_std = np.std(humidity)
correlation = covariance / (temp_std * humidity_std)

print("Correlation Cof: ", correlation)
print("-" * 40)

corr_matrix = np.corrcoef(temperature.flatten(), humidity.flatten())
print("np.corrcoef(): \n", corr_matrix)
```
