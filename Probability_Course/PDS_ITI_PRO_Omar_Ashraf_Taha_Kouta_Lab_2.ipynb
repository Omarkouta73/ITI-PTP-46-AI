# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

### True/Analytical PDF


## using `numpy`

```python
n_samples = 10000
X = np.random.normal(loc=0, scale=1, size=n_samples)
sample_mean_X = np.mean(X)
sample_variance_X = np.var(X)
sample_std_X = np.std(X, ddof=1)

x_range = np.linspace(-4, 4, 1000)
mu, sigma_squared = 0, 1
pdf_numpy = (1 / np.sqrt(2 * np.pi * sigma_squared)) * np.exp(-(x_range - mu)**2 / (2 * sigma_squared))
```

```python
#type your code here:
print(f"Sample Mean: {sample_mean_X:.4f}")
print(f"Sample Variance: {sample_variance_X:.4f}")

fig, axes = plt.subplots(figsize=(7, 5))

axes.hist(X, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black', label='Histogram')
axes.plot(x_range, pdf_numpy, 'r-', linewidth=2, label='True PDF')
axes.set_xlabel('x')
axes.set_ylabel('Probability Density')
axes.set_title('Standard Gaussian: Histogram vs Analytical PDF')
axes.legend()
axes.grid(True, alpha=0.3)
```

## using `scipy`

```python
#type your code here:
print(f"Sample Mean: {sample_mean_X:.4f}")
print(f"Sample Variance: {sample_variance_X:.4f}")

pdf_scipy = stats.norm.pdf(x_range, loc=0, scale=1)

fig, axes = plt.subplots(figsize=(7, 5))

axes.hist(X, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black', label='Histogram')
axes.plot(x_range, pdf_scipy, 'r-', linewidth=2, label='True PDF')
axes.set_title('Standard Gaussian: Histogram vs scipy PDF')
axes.legend()
axes.grid(True, alpha=0.3)

plt.show()
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
#type your code here:
cdf_scipy = stats.norm.cdf(x_range, loc=0, scale=1)

plt.figure(figsize=(7,5))

plt.hist(X, density=True, alpha=0.7, color='blue', label='Histogram')

plt.plot(x_range, pdf_scipy, color='red', linewidth=1.5, label='PDF')

plt.plot(x_range, cdf_scipy, color='orange', linewidth=1.5, label='CDF')

plt.xlabel('x')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()

```

e) Repeat for an binomial random variable with $n = 5$ and $p = 0.5$

```python
#type your code here:
n, p = 5, 0.5
B1 = np.random.binomial(n=n, p=p, size=n_samples)

sample_mean_B1 = np.mean(B1)
sample_variance_B1 = np.var(B1, ddof=1)
sample_std_B1 = np.std(B1, ddof=1)

print(f"Sample Mean: {sample_mean_B1:.4f}")
print(f"Sample Variance: {sample_variance_B1:.4f}")

k_range = np.arange(0, n + 1)

pmf_B1 = stats.binom.pmf(k_range, n=n, p=p)
cdf_B1 = stats.binom.cdf(k_range, n=n, p=p)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

counts, bins, patches = axes[0].hist(B1, bins=np.arange(-0.5, n + 1.5, 1), density=True, 
                                     alpha=0.7, color='steelblue', edgecolor='white', label='Histogram')
axes[0].stem(k_range, pmf_B1, linefmt='r-', markerfmt='ro', basefmt=' ', label='PMF')
axes[0].set_xticks(k_range)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].step(k_range, cdf_B1, 'b-', where='post', linewidth=2, label='CDF')
axes[1].stem(k_range, pmf_B1, linefmt='r-', markerfmt='ro', basefmt=' ', label='PMF')
axes[1].set_xticks(k_range)
axes[1].set_ylim([0, 1.1])
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
